```python
import urllib.request
from bs4 import BeautifulSoup
import numpy as np
import random
import time
url="https://www.flipkart.com/search?q=indian+men+dress&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="
count=0
Prices=[]
Ratings=[]
k2=0
for i in range(100):
  print(i)
  driver = webdriver.Firefox()
  url1=url+str(i+1)
  driver.get(url=url1)
  soup = BeautifulSoup(driver.page_source, 'html.parser')
  images = soup.find_all('img')
  print(len(images))
  p=0 #skip first 2 images as they are logos of Amazon
  for item in images:
      items=(item.get('src'))
      if "http" in items:
        if p<=2:
            p+=1
            continue
        k=urllib.request.urlretrieve(items, "images1/dress"+str(count+1)+".png")
        count+=1
        k2=0
  #for Price Extarction
  k=soup.find_all("div",attrs={"class":"_25b18c"})
  idiv="a-price-whole"
  for div in k:
    innerDivs= div.find_all("div",attrs={"class":"_30jeq3"})
    for idiv in innerDivs:
        Prices.append(idiv.get_text())
  driver.close()
    
    
print("Done")
```
